Count of all injuries sustained in the dataset by type.

```python
injury_counts = df['Injury_Type'].value_counts()
injury_counts
```

```python
plt.figure(figsize=(10,6))
injury_counts.plot(kind='bar', color='darkred')
plt.title('Frequency of Injury Types')
plt.ylabel('Number of Injuries')
plt.xlabel('Injury Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Set visual style.

```python
sns.set(style="whitegrid")
```

Frequency of injury types by count.

```python
plt.figure(figsize=(10,6))
injury_counts = df['Injury_Type'].value_counts()
sns.barplot(x=injury_counts.index, y=injury_counts.values, palette="coolwarm")
plt.title('Frequency of Injury Types', fontsize=16)
plt.ylabel('Number of Injuries', fontsize=12)
plt.xlabel('Injury Type', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Rehabilitation time by injury severity in a box plot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Injury_Severity', y='Rehabilitation_Time_weeks', data=df, palette="Set2")
plt.title('Rehabilitation Time by Injury Severity', fontsize=16)
plt.ylabel('Rehabilitation Time (weeks)', fontsize=12)
plt.xlabel('Injury Severity', fontsize=12)
plt.tight_layout()
plt.show()
```

Display the rehabilitation efficiency vs the rehabilitation time on a scatter plot.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Rehabilitation_Time_weeks', 
    y='Rehabilitation_Efficiency_Score', 
    hue='Injury_Severity', 
    data=df, 
    palette="Set1",
    s=100
)
sns.regplot(
    x='Rehabilitation_Time_weeks', 
    y='Rehabilitation_Efficiency_Score', 
    data=df, 
    scatter=False, 
    color='grey'
)
plt.title('Rehabilitation Efficiency vs Rehab Time', fontsize=16)
plt.ylabel('Rehabilitation Efficiency Score', fontsize=12)
plt.xlabel('Rehabilitation Time (weeks)', fontsize=12)
plt.legend(title='Injury Severity')
plt.tight_layout()
plt.show()
```

Display a heatmap of biomechanical metrics correlation for injury and performance.

```python
# Select the biomechanical metrics
biomech_metrics = [
    'knee_angle_deg',
    'jump_height_cm',
    'ankle_flexion_deg',
    'speed_m_s',
    'reaction_time_ms'
]

# Compute correlation matrix
corr_matrix = df[biomech_metrics].corr()

# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='coolwarm', 
    linewidths=0.5, 
    fmt=".2f",
    cbar_kws={'label': 'Correlation Coefficient'}
)
plt.title('Correlation of Biomechanical Metrics', fontsize=16)
plt.tight_layout()
plt.show()
```

Takeaways:
Players with high jump heights and better knee angles generally recover faster from lower-limb injuries. 
Reaction time and speed show weak correlations with eachother, suggesting indepedent physical traits.
